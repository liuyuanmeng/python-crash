````python
# 8.3 Boolean Filtering
# Let's return to the orders dataset for the rest of the challenges
# Use boolean filtering and DataFrame/DataSeries methods to solve the following challanges:
````

````python
# A. What is the mean profit of orders where the ship_mode is "Second Class"
filtered_orders_second_class = orders[orders['ship_mode'] == 'Second Class']
filtered_orders_second_class
````

````python
mean_profit = filtered_orders_second_class['profit'].mean()
mean_profit
````

````python
mask = (orders['ship_mode'] == 'Second Class') | (orders['ship_mode'] == 'First Class')
filtered_orders = orders[mask]
mask_profit = filtered_orders['profit'].mean()
mask_profit
````

````python
orders['profit'].mean()
````

````python
# B. Which product is the most commonly ordered on 2017-06-08 
most_commom = orders['order_date'] == '2017-06-08'
orders[most_commom]['product_id'].describe()
````

````python
# 8.4 Sorting and Filtering
````

````python
# A. What are the three most profitable orders purchased by customer PO-8865?
customer = orders['customer_id'] == 'PO-8865'
orders[customer].head(3)
````

````python
# B. What are the five orders of product_id TEC-HEW-10002304 with the highest quanitity?
product = orders[orders['product_id'] == 'TEC-HEW-10002304']
sorted_product = product.sort_values('quantity', ascending=False)
top_five_orders = sorted_product.head(5)
top_five_orders
````
